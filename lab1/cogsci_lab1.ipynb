8. Let's deal with the missing values. 

a) For the features EthnicGroup, ParentMaritalStatus and WklyStudyHours replace NaN values with the most common value.

b) For the feature ParentEduc replace NaN value with a string *no formal education*.

c) For the feature TestPrep replace NaN value with a string *none*.

d) For the features PracticeSport and IsFirstChild remove all the examples with NaN values. 

e) For the feature NrSiblings replace the NaN value with the median value.

f) Finally, for the feature TransportMeans replace the NaN value with a string *walking*.




```python
# a)

columns_to_fill = ["EthnicGroup", "ParentMaritalStatus", "WklyStudyHours"]
# mod je statistička funkcija koja vraća najčešću vrijednost u skupini
# mode() vraca DataFrame s jednim retkom u kojem su modovi stupaca
# da bi se dobila lista modova, moze se referencirati taj (prvi i jedini) redak pod implicitnim indeksom 0
# koristeci tu listu, mogu se popuniti NaN vrijednosti odgovarajucih stupaca
modes = df[columns_to_fill].mode().iloc[0]
df[columns_to_fill] = df[columns_to_fill].fillna(modes)

# b)

df["ParentEduc"].fillna("no formal education", inplace=True)

# c)

df["TestPrep"].fillna("none", inplace=True)

# d)

df.dropna(subset=["PracticeSport", "IsFirstChild"], inplace=True)
df = df.reset_index(drop=True)

# e)

median = df["NrSiblings"].median()
df["NrSiblings"].fillna(median, inplace=True)

# f)

df["TransportMeans"].fillna("walking", inplace=True)
```

10. The final step to preparing the dataset for classification is transforming the non-numeric variables to numeric variables. 
For features Gender and LunchType use OneHot Encoder, and for EthnicGroup, ParentEduc, TestPrep, ParentMaritalStatus, PracticeSport, IsFirstChild, TransportMeans and WklyStudyHours use Label Encoder. 

```python
encoder = OneHotEncoder(handle_unknown="ignore")
encoded_df = pd.DataFrame()
cols = ["Gender", "LunchType"]

# tranformira kategorije u boolean stupce
for column in cols:
    enc_column = encoder.fit_transform(df[[column]]) # encode DataFrame's column
    # spremi kodirane stupce u novi DataFrame i spoji ga s izvornim df-om
    generated_columns_df = pd.DataFrame.sparse.from_spmatrix(enc_column, columns=encoder.get_feature_names_out([column]))
    df = pd.concat([df, generated_columns_df], axis=1)
```

```python
df
```

```python
le = LabelEncoder()
label_cols=["EthnicGroup", "ParentEduc", "TestPrep", "ParentMaritalStatus", "PracticeSport", "IsFirstChild", "TransportMeans", "WklyStudyHours"]

# transformira kategorije u nummericke vrijednosti
for label in label_cols:
    le.fit(df.loc[:,label])
    df.loc[:,label] = le.transform(df.loc[:,label])
```
